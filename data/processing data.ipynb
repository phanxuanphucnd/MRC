```python
import json
import random
import numpy as np
import pandas as pd
```

```python
with open('./uit-visquad/train.json', 'r+', encoding='utf-8') as f:
    data = json.load(f)
```

```python
data_ = data['data']
len(data_)
```

### Split by paragraphs

```python
NUM_TRAIN_PRAGRAPHS = 125
NUM_DEV_PARAGRAPHS = 13

list_paragraphs = []

for i in range(len(data_)):
    list_paragraphs.append(data_[i])
```

```python
SEED = 123
kfolds = 10
random.seed(SEED)


indexes = [i for i in range(len(list_paragraphs))]

for k in range(kfolds):
    train_data = {
        'version': f'viquad2_training_set_{k}',
        'data': []
    }
    dev_data = {
        'version': f'viquad2_dev_set_{k}',
        'data': []
    }
    
    train_ids = random.sample(indexes, k=NUM_TRAIN_PRAGRAPHS)
    dev_ids = [i for i in indexes if i not in train_ids]
    print(f"Train ids: ", len(train_ids), train_ids)
    print(f"Dev ids: ", len(dev_ids), dev_ids)
    
    for id in train_ids:
        train_data['data'].append(list_paragraphs[id])
        
    for id in dev_ids:
        dev_data['data'].append(list_paragraphs[id])
        
    with open(f'./uit-visquad/kfolds/train-{k}.json', 'w', encoding='utf-8-sig') as f:
        json.dump(train_data, f, ensure_ascii=False, indent=4)
        
    with open(f'./uit-visquad/kfolds/dev-{k}.json', 'w', encoding='utf-8') as f:
        json.dump(dev_data, f, ensure_ascii=False, indent=4)
```

```python
data_ = data['data']

for i in range(len(data_)):
    paragraphs = data_[i]['paragraphs']
    for p in range(len(paragraphs)):
        qas = paragraphs[p]['qas']
        for q in range(len(qas)):
            qas[q]["answers"] = []
            qas[q]["is_impossible"] = False
```

```python
with open(f'./uit-visquad/private_test.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
data_dict = {
#     'title': [],
#     'context': [],
#     'question': [],
    'text': [], 
    'label': []
}

data_ = data['data']

MAX = 0
MAX_CONTEXT = 0
MAX_QUESTION = 0

for i in range(len(data_)):
    title = data_[i]['title']
    paragraphs = data_[i]['paragraphs']
    for par in paragraphs:
        context = par['context']
        if len(context.split()) > MAX_CONTEXT:
            MAX_CONTEXT = len(context.split())
            
        qas = par['qas']
        for q in qas:
            question = q['question']
            if len(question.split()) > MAX_QUESTION:
                MAX_QUESTION = len(question.split())
            label = 0 if not q['is_impossible'] else 1
            
            text = question.strip() + ' ' + context.strip()
            if len(text.split()) > MAX:
                MAX = len(text.split())
            
#             data_dict['title'].append(title.lower())
#             data_dict['context'].append(context.lower())
#             data_dict['question'].append(question.lower())
            data_dict['text'].append(text.lower())
            data_dict['label'].append(label)
```

```python
data_df = pd.DataFrame.from_dict(data_dict)
print("MAX = ", MAX, MAX_CONTEXT, MAX_QUESTION)

print(len(data_df))
data_df.head(5)
```

```python
data_df.to_csv("./uit-visquad/ptest.csv", encoding='utf-8', index=False)
```
